# librosa

librosa is a python package for Audio analysis.

The audio signal (and therefore, the audio file) consists of **signal_strength** and **sample_rate**. **signal_strength** is simply a vector of real numbers each of which is sampled at the **sample_rate**.

```python
%matplotlib inline

import librosa
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read a sample audio signal and plot it

If the **sample_rate** is 22050 (per second) and the audio length in time is 10 min, the number of samples will be 22050 \* 10 \* 60 (== 13.23 million).

```python
# Load the example clip
y, sr = librosa.load(librosa.util.example_audio_file(), duration=10)
print(type(y), y.shape, sr)
print('Duration: {}'.format(librosa.get_duration(y=y, sr=sr)))

# Plot the signal
plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
librosa.display.waveplot(y, sr=sr)
```

## MFCC features

Now that we know how to load an audio signal using `librosa`, let's try to get the MFCC features (which are commonly used for speech recognition) from our signal.

```python
mfcc = librosa.feature.mfcc(y=y, sr=sr)
print(type(mfcc), mfcc.shape)

plt.figure(figsize=(10, 6))
librosa.display.specshow(mfcc, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
```
